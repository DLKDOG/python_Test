```python
import matplotlib.pyplot as plt

x=range(100)
y=range(100)
plt.plot(x,y)
```

```python
fig,ax=plt.subplots()

X_1=range(100)
Y_1=[np.cos(value) for value in x]

ax.plot(X_1,Y_1)
ax.set(title='cos graph',xlabel="X",ylabel='Y' );

plt.show()
```

```python
x=np.linspace(-1,1,100)
y_1=np.sin(x)
y_2=np.cos(x)
y_3=np.tan(x)
y_4=np.exp(x)

fig, ax=plt.subplots(2,2)

ax[0,0].plot(x,y_1)
ax[0,1].plot(x,y_2)
ax[1,0].plot(x,y_3)
ax[1,1].plot(x,y_4)
```

```python
X_1 = range(100)
Y_1 = [value for value in X_1]

X_2 = range(100)
Y_2 = [value +100 for value in X_2]

plt.plot(X_1,Y_1,color="#000000")
plt.plot(X_2,Y_2,c='c')

plt.show()

```
