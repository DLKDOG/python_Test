```python
import matplotlib.pyplot as plt

# 초기 값 설정
x_new = 10  # 초기 x 값
derivative = []  # 경사 값(파라미터 업데이트 시점 기록)
y = []  # f(x) 값 기록
learning_rate = 0.1  # 학습률 설정

# 경사하강법 수행
for i in range(100):  # 100번 반복
    old_value = x_new
    grad = 2 * old_value  # f'(x) = 2x에 해당하는 기울기 계산
    x_new = old_value - learning_rate * grad  # x 값 업데이트
    
    derivative.append(x_new)  # x 값 업데이트 기록
    y.append(x_new ** 2)  # f(x) 값 기록

# 첫 번째 그래프: x 값의 변화
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(range(len(derivative)), derivative, color='red', label='x values after updates')
plt.title("x Value Updates")
plt.xlabel("Iteration")
plt.ylabel("x Value")
plt.grid(True)

# 두 번째 그래프: f(x) 값의 변화
plt.subplot(1, 2, 2)
plt.plot(range(len(y)), y, color='blue', label='f(x) values')
plt.title("f(x) Value Updates")
plt.xlabel("Iteration")
plt.ylabel("f(x) Value")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) = x^2 and its derivative f'(x) = 2x
def f(x):
    return x**2

def df(x):
    return 2 * x

# Set the initial x value and the learning rate
x = 3
learning_rate = 0.1

# Prepare lists to keep track of x values and function values for visualization
x_values = [x]
f_values = [f(x)]

# Perform a few iterations of gradient descent
iterations = 5
for _ in range(iterations):
    gradient = df(x)
    x = x - learning_rate * gradient
    x_values.append(x)
    f_values.append(f(x))

# Plotting the function and the points showing gradient descent steps
x_range = np.linspace(-1, 3.5, 100)
y_range = f(x_range)

plt.figure(figsize=(8, 6))
plt.plot(x_range, y_range, label=r'$f(x) = x^2$', color='blue')
plt.scatter(x_values, f_values, color='red', zorder=5, label="Gradient Descent Steps")
plt.plot(x_values, f_values, color='red', linestyle='dashed', alpha=0.6)
for i, (x_val, f_val) in enumerate(zip(x_values, f_values)):
    plt.text(x_val, f_val + 0.5, f"Step {i}", fontsize=12)

plt.title("Gradient Descent on $f(x) = x^2$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) = x^2 and its derivative f'(x) = 2x
def f(x):
    return x**2

def df(x):
    return 2 * x

# Set the initial x value and the learning rate
x = 3
learning_rate = 0.1
tolerance = 1e-5  # 최소값에 얼마나 가까워질지를 설정 (0에 가까워지는 기준)

# Prepare lists to keep track of x values and function values for visualization
x_values = [x]
f_values = [f(x)]

# Perform gradient descent until x is close to 0
iteration = 0
while abs(x) > tolerance:  # x가 0에 가까워질 때까지 반복
    gradient = df(x)
    x = x - learning_rate * gradient
    x_values.append(x)
    f_values.append(f(x))
    iteration += 1
    
# 최종 x 값과 반복 횟수 계산    
final_x_value = x_values[-1]
iteration_count = iteration


# Plotting the function and the points showing gradient descent steps
x_range = np.linspace(-1, 3.5, 100)
y_range = f(x_range)

plt.figure(figsize=(8, 6))
plt.plot(x_range, y_range, label=r'$f(x) = x^2$', color='blue')
plt.scatter(x_values, f_values, color='red', zorder=5, label="Gradient Descent Steps")
plt.plot(x_values, f_values, color='red', linestyle='dashed', alpha=0.6)
for i, (x_val, f_val) in enumerate(zip(x_values, f_values)):
    plt.text(x_val, f_val + 0.5, f"Step {i}", fontsize=12)

plt.title(f"Gradient Descent on $f(x) = x^2$ (Steps: {iteration})")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()

# Display the results
iteration_count, final_x_value
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random

def gen_data(numPoints, bias, variance):
    x = np.zeros(shape=(numPoints, 2))  # (2) 데이터 x의 첫 번째 열에는 1을 넣기 위한 초기화
    y = np.zeros(shape=numPoints)       # (4) y 데이터를 위한 초기화
    
    for i in range(0, numPoints):
        x[i][0] = 1  # (2) 데이터 x의 첫 번째 열은 1로 채워집니다.
        x[i][1] = i  # (3) 데이터 x의 두 번째 열은 0부터 시작하는 증가 값입니다.
        y[i] = (i + bias) + random.uniform(0, 1)*variance  # (4) y 데이터에 bias와 노이즈를 추가하여 생성
    return x, y
```
